# Explorando o Apache Iceberg
O Iceberg é focado em performance para tabelas gigantescas e organização rigorosa através de catálogos.

```
from pyspark.sql import SparkSession

# Configuração para Iceberg com Catálogo Local (Hadoop)
spark = SparkSession.builder.appName("Estudo_Iceberg") \
    .config("spark.jars.packages", "org.apache.iceberg:iceberg-spark-runtime-3.5_2.12:1.5.0") \
    .config("spark.sql.extensions", "org.apache.iceberg.spark.extensions.IcebergSparkSessionExtensions") \
    .config("spark.sql.catalog.local", "org.apache.iceberg.spark.SparkCatalog") \
    .config("spark.sql.catalog.local.type", "hadoop") \
    .config("spark.sql.catalog.local.warehouse", "data/iceberg_warehouse") \
    .getOrCreate()

print("Sessão Spark com Apache Iceberg pronta!")
```

```
# Criando uma base de dados e uma tabela no Iceberg
spark.sql("CREATE DATABASE IF NOT EXISTS local.db_teste")

data = [("Engenharia", 1), ("Dados", 2)]
df = spark.createDataFrame(data, ["palavra", "valor"])

# No Iceberg, preferimos usar o comando writeTo
df.writeTo("local.db_teste.tabela_v1").createOrReplace()

print("Tabela Iceberg criada no catálogo 'local'")
spark.table("local.db_teste.tabela_v1").show()
```

```
# Verificando os metadados (O diferencial do Iceberg)
print("Explorando os arquivos de log e metadados na pasta data/iceberg_warehouse...")
spark.sql("SELECT * FROM local.db_teste.tabela_v1.snapshots").show()
```
